## Como interpretar

- Overlay: verde = mascara real, rojo = prediccion. Si coinciden, se ve mezcla amarillo/rojiza sobre la lesion.
- Mapa de errores: verde = acierto lesion, rojo = falso positivo, azul = falso negativo.
- Si aparece mucho rojo, el modelo sobresegmenta. Si aparece mucho azul, se queda corto y pierde lesion.
